```python
def check_possible(row, col, _board):
    possible = [1,2,3,4,5,6,7,8,9]

    # in same row
    for num in possible.copy():
        if num in _board[row]:
            possible.remove(num)

    # in same column
    for num in possible.copy():
        if num in [x[col] for x in _board]:
            possible.remove(num)

    # in same field
    if row < 3:
        rows = _board[:3].copy()
    elif row < 6:
        rows = _board[3:6].copy()
    else:
        rows = _board[6:].copy()

    if col < 3:
        rows = [x[:3] for x in rows]
    elif col < 6:
        rows = [x[3:6] for x in rows]
    else:
        rows = [x[6:] for x in rows]

    for num in possible.copy():
        for row in rows:
            if num in row:
                possible.remove(num)
    
    return possible

board = [
    [1, 2, 3, 4, 5, 6, 7, 8, 9],
    [4, 6, 9, 9, 9, 9, 2, 3, 5],
    [9, 9, 9, 7, 3, 8, 6, 6, 6],
    [9, 9, 8, 9, 9, 9, 9, 9, 7],
    [9, 9, 8, 9, 9, 9, 9, 4, 1],
    [9, 9, 7, 9, 8, 4, 6, 6, 6],
    [9, 9, 8, 8, 9, 9, 4, 9, 8],
    [9, 9, 8, 8, 9, 9, 3, 5, 8],
    [0, 4, 8, 8, 7, 1, 6, 6, 6],
]

check_possible(8, 0, board)
```
